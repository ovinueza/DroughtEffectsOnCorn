```python
import pandas as pd
```

```python
price = pd.read_csv("price.csv")
price.head()
```

```python
list(price.groupby(["Year"])["Values"].mean())
```

```python
df = pd.DataFrame(price.groupby(["Year","State"])["Values"].sum())\
.reset_index().set_index("Year").pivot(columns = "State", values = "Values")

df_price = df[["IOWA","ILLINOIS","NEBRASKA","MINNESOTA","INDIANA",\
               "SOUTH DAKOTA","KANSAS","OHIO","WISCONSIN","MISSOURI",\
               "NORTH DAKOTA","MICHIGAN","KENTUCKY","TEXAS","COLORADO",\
               "PENNSYLVANIA","ARKANSAS","TENNESSEE","NORTH CAROLINA","MISSISSIPPI"]]
df_price.head()
```

```python
production = pd.read_csv("production.csv")
production.head()
```

```python
list(production.groupby(["Year"])["Value"].sum())
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
df = pd.DataFrame(production.groupby(["Year","State"])["Value"].sum())\
.reset_index().set_index("Year").pivot(columns = "State", values = "Value")

df_production = df[["IOWA","ILLINOIS","NEBRASKA","MINNESOTA","INDIANA",\
               "SOUTH DAKOTA","KANSAS","OHIO","WISCONSIN","MISSOURI",\
               "NORTH DAKOTA","MICHIGAN","KENTUCKY","TEXAS","COLORADO",\
               "PENNSYLVANIA","ARKANSAS","TENNESSEE","NORTH CAROLINA","MISSISSIPPI"]]
df_production.head()
```
