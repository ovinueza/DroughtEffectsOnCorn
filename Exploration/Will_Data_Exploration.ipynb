```python
import pandas as pd
```

```python
df = pd.read_csv("csv_files/Corn_forFuelAlcohol_Usage_National_MillType_2014-2019.csv")
df.head()
```

```python
df = df[["Period","Domain Category","Value"]]
df.head()
```

```python
type(df["Value"][0])
```

```python
df["Converted_Value"] = df["Value"].apply(lambda x: int(x.replace(',','')))
df.head()
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
pd.DataFrame(df.groupby(["Domain Category"])["Converted_Value"].mean())
```
